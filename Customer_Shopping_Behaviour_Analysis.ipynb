```python
#removing null values in the review rating column by replacing each category item with the median of that category to not include bias
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#assigning a new column >> feature engineering
label = ['Yound-Adults','Adults','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=label)
```

```python
df[['age','age_group']].head(5)
```

```python
#another feature purchase_frequency_days
#convert the text form to numbers
# df['frequency_of_purchases'] = df['frequency_of_purchases'].str.strip().str.title()

frequency_mapping = {
    'Fortnightly': 14, 
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}


df['purchase_frequency_days']= df['frequency_of_purchases'].map(frequency_mapping)
```
